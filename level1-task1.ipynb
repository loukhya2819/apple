```python
# Task 1 - Level 1: Student Score Analysis

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("/student_scores.csv")

# Show first 5 records
print("ðŸ”¹ First 5 Records:")
print(df.head())

# Basic info
print("\nðŸ”¹ Dataset Info:")
print(df.info())

# Descriptive statistics
print("\nðŸ”¹ Descriptive Statistics:")
print(df.describe())

# Check for missing values
print("\nðŸ”¹ Missing Values:")
print(df.isnull().sum())

# Correlation heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(df.drop("Name", axis=1).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Subjects & Attendance")
plt.show()

# Plot subject-wise scores
plt.figure(figsize=(10, 5))
df.drop("Name", axis=1).plot(kind="box")
plt.title("Score Distribution per Subject")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```
